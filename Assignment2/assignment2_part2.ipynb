## Part 2 Laplace approximation for a simple neural network

**Task 2.1: Generate $S = 100$ samples from the prior $p(w)$ and plot the corresponding functions
$f(x)$ for $x \in [0, 10]$ on top of a scatter plot of the data.**

```python
sigmoid = lambda x: 1./(1+jnp.exp(-x))

tau = 2
beta = 4

x_data = jnp.array([9.589, 7.375, 4.647, 2.501, 2.538, 6.783, 4.294, 5.111, 0.130, 0.783])
y_data = jnp.array([3.032, 3.349, 2.906, 2.126, 1.538, 2.787, 3.078, 2.993, 0.828, -0.331])

def f(w,x):
    return w[1]*sigmoid(x+w[0])

w_sample = norm.rvs(size=(100,2), loc=0, scale=tau)

x = jnp.linspace(0,10,1000)
# plot f(x) for each sample of w on top of data
fig, ax = plt.subplots()
ax.scatter(x_data, y_data,label='Data')
for w in w_sample[:-1]:
    ax.plot(x, f(w,x), color='black', alpha=0.3)
ax.plot(x, f(w_sample[-1],x), color='black', alpha=0.3, label=r'$f(x|\mathbf{w}) = w_1\sigma(x+w_0) $'+'\n' + r'for each $\mathbf{w} = [w_0, w_1]$')
ax.set_ylim(-2,4)
ax.set_xlim(0,10)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$ / $f(x)$')
ax.legend(loc='lower right')
# title
plt.title(r'Two parameter network $f(x|\mathbf{w})$ from 100 $\mathbf{w}$ samples')
plt.tight_layout()
plt.show() 
```

The joint density for $(y,\mathbf{w})$ according to the probabilistic model in eq. (11)-(13) is given
$$
\begin{align}\tag{14}
p(\boldsymbol{y}, \boldsymbol{w})&=p(\boldsymbol{w}) \prod_{n=1}^N p\left(y_n \mid x_n, \boldsymbol{w}\right)\\
&=\mathcal{N}\left(w_0 \mid 0, \tau^2\right) \mathcal{N}\left(w_1 \mid 0, \tau^2\right) \prod_{n=1}^N \mathcal{N}\left(y_n \mid w_1 \sigma\left(x_n+w_0\right), \beta^{-1}\right)\notag
\end{align}
$$

**Task 2.2: What prevents us from using the equations in section 3.3 in Murphy1 to compute
the exact posterior distribution of the parameters given the data analytically for the system
in eq. (14) ?**

The joint distribution will also be a gaussian, however, the mean of the likelihood gaussian, $f(x)$, is non-linear, hence, it is not possible to derive a Gaussian distribution analytically, through the equations of section 3.3 *Linear Gaussian systems* in Murphy1.

**Task 2.3: Implement a python function for evaluating the logarithm of the joint density in
eq. (14). Report the numerical value of $\text{log}\ p(y,w)$ for the dataset given above when $w_0 = w_1 = 0$.**

$$
\begin{align}
\log p(\mathbf{y},\mathbf{w}) &= \log \left[ \mathcal{N}\left(w_0 \mid 0, \tau^2\right) \mathcal{N}\left(w_1 \mid 0, \tau^2\right) \prod_{n=1}^N \mathcal{N}\left(y_n \mid w_1 \sigma\left(x_n+w_0\right), \beta^{-1}\right)\right]\tag{2.3a}\\
&=
- \frac{2+N}{2}\log (2\pi) - 2\log\tau - \frac{w_0^2+w_1^2}{2\tau^2} + \frac{N}{2}\log \beta - \sum_{n=1}^N \beta\frac{(y_n - w_1 \sigma\left(x_n+w_0\right))^2}{2}\tag{2.3b}
\end{align}
$$

```python
# Testing the analytical expression is derived correctly
def log_joint_from_dist(w, y, x):
    return norm.logpdf(w[0], loc=0, scale=tau) + norm.logpdf(w[1], loc=0, scale=tau) + jnp.sum(norm.logpdf(y, loc=f(w,x), scale=1/jnp.sqrt(beta)))

def log_joint(w, y, x):
    return -(2+len(y))/2*jnp.log(2*jnp.pi) - 2*jnp.log(tau) - (w[0]**2+w[1]**2)/(2*tau**2) + len(y)/2*jnp.log(beta) - jnp.sum(beta*(y-f(w,x))**2)/2

w0 = 0
w1 = 0
print('from log-normal distributions (2.3a):',abs(log_joint_from_dist([w0,w1], y_data, x_data)))
print('from analytical expression (2.3b):',abs(log_joint([w0,w1], y_data, x_data)))

print(f'The two functions differ with {abs(log_joint_from_dist([w0,w1], y_data, x_data) - log_joint([w0,w1], y_data, x_data)):.6f}.')
```

**Task 2.4: Determine the expression for the gradient and Hessian of the logarithm of the joint
density wrt. $\mathbf{w}$.**

**The gradient:**

$$\begin{align*}
\nabla_{\mathbf{w}} \log p(\mathbf{y}, \mathbf{w}) &= 
\nabla_{\mathbf{w}} \left( \underbrace{- \frac{2+N}{2}\log (2\pi) - 2\log\tau + \frac{N}{2}\log \beta}_{\text{constant wrt. }\mathbf{w}} - \frac{1}{2\tau^2}\mathbf{w}^T\mathbf{w} - \sum_{n=1}^N \beta\frac{\left(y_n - w_1 \sigma\left(x_n+w_0\right)\right)^2}{2} \right) \\
&=
-\frac{1}{\tau^2} -\frac{\beta}{2} \sum_{n=1}^N \nabla_{\mathbf{w}}\left[y_n^2 - 2 y_n w_1 \sigma\left(x_n+w_0\right) + \left(w_1 \sigma\left(x_n+w_0\right)\right)^2\right]
\end{align*}
$$
To make the derivation simpler, we'll now differentiate the summation term with respect to the two components of $\mathbf{w}$ separately. We'll also denote $\sigma(x+w_0)=\sigma$ and remember that $\partial_x \sigma(x+y) = \sigma(x+y)(1-\sigma(x+y))=\sigma'$.
$$
\begin{align*}
\frac{\partial}{\partial w_0} &\left[ y_n^2 - 2 y_n w_1 \sigma + w_1^2\sigma^2 \right] = - 2 y_n w_1\sigma'+ 2w_1^2\sigma\sigma'

\\
\frac{\partial}{\partial w_1} &\left[  y_n^2 - 2 y_n w_1 \sigma + w_1^2\sigma^2 \right] =
 - 2 y_n\sigma + 2 w_1 \sigma^2
\end{align*}$$

The gradient then becomes
$$
\nabla_{\mathbf{w}} \log p(\mathbf{y}, \mathbf{w})=
-\frac{1}{\tau^2}\mathbf{w} -\beta \sum_{n=1}^N
\begin{pmatrix} - y_n w_1\sigma'+ w_1^2\sigma\sigma' \\ - y_n\sigma + w_1 \sigma^2 \end{pmatrix}
$$

plugging in the two expressions, we get

$$
\begin{align*}
\nabla_{\mathbf{w}} \log p(\mathbf{y}, \mathbf{w}) &=
-\frac{w_0+w_1}{\tau^2} -\beta \sum_{n=1}^N w_1\sigma^2 + w_1^2\sigma\sigma' - w_1y_n\sigma' - y_n\sigma\\
&=
-\frac{w_0+w_1}{\tau^2} -\beta \sum_{n=1}^N w_1\sigma^2 + w_1^2\sigma\sigma(1-\sigma) - w_1y_n\sigma(1-\sigma) - y_n\sigma\\
&=
-\frac{w_0+w_1}{\tau^2} -\beta \sum_{n=1}^N  - (1+w_1)y_n\sigma + (w_1 + w_1y_n + w_1^2)\sigma^2  -w_1^2\sigma^3
\end{align*}
$$


```python
### testing the gradient with point on plot
def gradient(w0, w1, y, x):

    sigma = sigmoid(x+w0)

    return -(w0+w1)/(tau**2) - beta*jnp.sum(-(1+w1)*y*sigma+(w1+w1*y+w1**2)*sigma**2-w1**2*sigma**3)

def grad(w0, w1, y, x):

    sigma = sigmoid(x+w0)
    sigma_m = sigma*(1-sigma)

    return -(w0+w1)/(tau**2) - beta*jnp.sum( w1*sigma**2 + w1**2*sigma*sigma_m -w1*y*sigma_m-y*sigma)
print('Gradient:',gradient(-2.20005,3.102,y_data,x_data))
print('Grad:',grad(-2.20005,3.102,y_data,x_data))
```

**The Hessian:**
$$ \begin{align*}
\mathcal{H}(\textbf{w}) = \sum_{i,j} \frac{\partial^2}{\partial w_i \partial w_j} \log p(\mathbf{y}, \mathbf{w}) &= \nabla_{\mathbf{w}} \left[-\frac{1}{\tau^2}\mathbf{w} -\beta \sum_{n=1}^N
\begin{pmatrix} - y_n w_1\sigma'+ w_1^2\sigma\sigma' \\ - y_n\sigma + w_1 \sigma^2 \end{pmatrix} \right]\\
\end{align*}$$

The partial derivatives:
$$
\begin{align*}
-\frac{1}{\tau^2}-\beta \sum_{n=1}^N\frac{\partial}{\partial w_0}  \begin{pmatrix} - y_n w_1\sigma'+ w_1^2\sigma\sigma' \\ - y_n\sigma + w_1 \sigma^2 \end{pmatrix}
&=
-\frac{1}{\tau^2}-\beta \sum_{n=1}^N\begin{pmatrix} - w_1y_n\sigma'' + w_1^2\sigma'\sigma' +w_1^2\sigma\sigma''  \\ - y_n \sigma' + 2w_1\sigma\sigma' \end{pmatrix}
\\

\\

-\frac{1}{\tau^2}-\beta \sum_{n=1}^N\frac{\partial}{\partial w_1} \begin{pmatrix} - y_n w_1\sigma'+ w_1^2\sigma\sigma' \\ - y_n\sigma + w_1 \sigma^2 \end{pmatrix}
&=
-\frac{1}{\tau^2}-\beta \sum_{n=1}^N\begin{pmatrix}  - y_n\sigma'+ 2w_1\sigma\sigma' \\ \sigma^2 \end{pmatrix}
\end{align*}$$

Plugging in the expressions for the partial derivatives, we get the hessian
$$
\mathcal{H}(\textbf{w})

\begin{pmatrix} - w_1y_n\sigma'' + w_1^2\sigma'\sigma' +w_1^2\sigma\sigma''  & - y_n \sigma' + 2w_1\sigma\sigma'\\
- y_n\sigma'+ 2w_1\sigma\sigma' h  \sigma^2
\end{pmatrix}

$$



```python
def Hessian(w0, w1, y, x):
    sigma = sigmoid(x+w0)
    sigma_m = sigma*(1-sigma)
    sigma_mm = sigma_m*(1-2*sigma)

    return -2/(tau**2) - beta*jnp.sum(4*w1*sigma_m*sigma + w1**2*sigma_m**2+w1**2*sigma*sigma_mm
                                       - w1*y*sigma_mm - 2*y*sigma_m + sigma**2)

print('Hessian:',Hessian(-2.20005,3.102,y_data,x_data))
```

**Task 2.5: Determine the Laplace approximation $q(w)$ of the posterior distribution such that
$p(w|y) \approx q(w)$.**

The Laplace approximation assumes that we can fit a gaussian to the posterior distribution locally at the MAP with variance corresponding to the inverse, negative hessian of the logaritmic joint distribution.
$$
p(\mathbf{w}|\mathbf{y})\approx q(\mathbf{w})=\mathcal{N}\left(\mathbf{w}|\mathbf{w}_{\text{MAP}},\left[-\nabla^2_{\mathbf{w}}\log p(\mathbf{w},\mathbf{y})\big\rvert_{\mathbf{w}=\mathbf{w}_{\text{MAP}}} \right]^{-1} \right)
$$

We can find the MAP estimate by taking the gradient of the log joint distribution and setting it equal to zero. This is called MAP estimation.
It builds on the idea behind MLE, but instead of maximizing just the likelihood, MAP maximizes the posterior.
$$
\begin{align*}
\nabla_{\mathbf{w}} \log p(\mathbf{y}, \mathbf{w}) &= -\frac{w_0+w_1}{\tau^2} -\frac{\beta}{2} \sum_{n=1}^N \nabla_{\mathbf{w}}\left[y_n^2 - 2 y_n w_1 \sigma\left(x_n+w_0\right) + \left(w_1 \sigma\left(x_n+w_0\right)\right)^2\right] = 0\\
\Leftrightarrow  w_0 + w_1 &= -\tau^2\frac{\beta}{2} \sum_{n=1}^N \nabla_{\mathbf{w}}\left[y_n^2 - 2 y_n w_1 \sigma\left(x_n+w_0\right) + \left(w_1 \sigma\left(x_n+w_0\right)\right)^2\right]
\end{align*}
$$
which leads to
$$
\begin{align*}
-\tau^2\frac{\beta}{2} \sum_{n=1}^N\frac{\partial}{\partial w_0} &\left[ y_n^2 - 2 y_n w_1 \sigma + w_1^2\sigma^2 \right] =-\tau^2\frac{\beta}{2} \sum_{n=1}^N
 - 2 y_n w_1\sigma'+ 2w_1^2\sigma\sigma' = w_0

\\
-\tau^2\frac{\beta}{2} \sum_{n=1}^N\frac{\partial}{\partial w_1} &\left[  y_n^2 - 2 y_n w_1 \sigma + w_1^2\sigma^2 \right] =-\tau^2\frac{\beta}{2} \sum_{n=1}^N
 - 2 y_n\sigma + 2 w_1 \sigma^2 = w_1
\end{align*}$$

Hence, the Laplace approximation becomes:
$$
q(\mathbf{w}) = \mathcal{N}\left(\mathbf{w} \mid \mathbf{w}_{\text{MAP}}, \left[-\nabla^2_{\mathbf{w}} \log p(\mathbf{w}, \mathbf{y}) \big|_{\mathbf{w} = \mathbf{w}_{\text{MAP}}} \right]^{-1} \right)
$$

where $$ \mathbf{w}_{\text{MAP}} = [w_0, w_1]^T $$ is the value of the parameters that maximizes the posterior distribution, i.e., the point that solves the system of equations above, obtained by setting the gradient of the log joint distribution to zero.


**Task 2.6: Create a 2D contour plot of posterior distribution of the parameters. Plot the
contours of the Laplace approximation on top of the plot to validate your results from the
previous tasks.**

```python

# Forward model
def f_grid(w0, w1, x):
    return w1 * sigmoid(x[:, None, None] + w0)

# Log joint probability
def log_joint(w0, w1, y, x):
    N = len(y)
    return -(2 + N) / 2 * jnp.log(2 * jnp.pi) - 2 * jnp.log(tau) - (w0**2 + w1**2) / (2 * tau**2) + N / 2 * jnp.log(beta) - beta * jnp.sum((y[:, None, None] - f_grid(w0, w1, x))**2) / 2

# Laplace approximation (Gaussian)
def log_q(w, w_star, H):
    diff = w - w_star
    return -0.5 * diff.T @ jnp.linalg.inv(H) @ diff - 0.5 * jnp.log(jnp.linalg.det(2 * jnp.pi * H))

# Plotting function
def plot_posterior_and_laplace(y, x, w0_range, w1_range, num_points=100):
    w0_vals = jnp.linspace(w0_range[0], w0_range[1], num_points)
    w1_vals = jnp.linspace(w1_range[0], w1_range[1], num_points)
    w0_grid, w1_grid = jnp.meshgrid(w0_vals, w1_vals)

    # Compute log joint over grid
    log_joint_vals = jnp.vectorize(lambda w0, w1: log_joint(w0, w1, y, x))(w0_grid, w1_grid)

    # MAP estimate
    max_idx = jnp.unravel_index(jnp.argmax(log_joint_vals), log_joint_vals.shape)
    w0_star = w0_vals[max_idx[1]]
    w1_star = w1_vals[max_idx[0]]
    w_star = jnp.array([w0_star, w1_star])

    # Hessian at MAP
    hessian_func = hessian(lambda w: log_joint(w[0], w[1], y, x))
    H = -hessian_func(w_star)

    # Laplace approx over grid
    log_q_vals = jnp.vectorize(lambda w0, w1: log_q(jnp.array([w0, w1]), w_star, H))(w0_grid, w1_grid)

    # Plot
    plt.figure(figsize=(10, 8))

    # Posterior contours
    contour1 = plt.contour(w0_grid, w1_grid, jnp.exp(log_joint_vals), levels=15, colors='blue')
    plt.clabel(contour1, inline=True, fontsize=8)
    
    # Laplace contours
    contour2 = plt.contour(w0_grid, w1_grid, jnp.exp(log_q_vals), levels=15, colors='red', linestyles='dashed')
    plt.clabel(contour2, inline=True, fontsize=8)

    # MAP estimate point
    plt.plot(w0_star, w1_star, 'ko', markersize=8, label='MAP estimate $\\mathbf{w}^*$')

    # Axis and legend
    plt.xlabel('$w_0$', fontsize=12)
    plt.ylabel('$w_1$', fontsize=12)
    plt.title('Posterior Distribution and Laplace Approximation', fontsize=14)

    # Custom legend
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], color='blue', label='Posterior Distribution'),
        Line2D([0], [0], color='red', linestyle='--', label='Laplace Approximation'),
        Line2D([0], [0], marker='o', color='k', label='MAP Estimate', markersize=8, linestyle='None')
    ]
    plt.legend(handles=legend_elements, loc='upper left', fontsize=10)

    # Adjust plot limits for clarity
    padding = 0.5
    plt.xlim(w0_range[0] - padding, w0_range[1] + padding)
    plt.ylim(w1_range[0] - padding, w1_range[1] + padding)

    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()


x = jnp.linspace(-3, 3, 10)
y = sigmoid(x + 1) + 0.1 * jnp.array(np.random.normal(size=10))

plot_posterior_and_laplace(y, x, w0_range=(-3, 3), w1_range=(-5, 6))
```

The Laplace approximation assumes the posterior is Gaussian around the MAP, but the true posterior is often non-Gaussian due to the nonlinear sigmoid in the model and limited data. This causes mismatches, especially in regions far from the MAP.


**Task 2.7: Generate and plot $S = 100$ samples from the approximate posterior distribution. Plot
the corresponding function $f(x)$ for $x \in [0, 10]$ on top of a scatter plot of the data.**

```python
import jax.numpy as jnp
from jax import hessian
import matplotlib.pyplot as plt
import numpy as np

log_joint_grid = jnp.vectorize(lambda w0, w1: log_joint(w0, w1, y_data, x_data))(w0_grid, w1_grid)
max_idx = jnp.unravel_index(jnp.argmax(log_joint_grid), log_joint_grid.shape)
w0_star = w0_range[max_idx[1]]
w1_star = w1_range[max_idx[0]]
w_star = jnp.array([w0_star, w1_star])

# Hessian and covariance
hess = hessian(lambda w: log_joint(w[0], w[1], y_data, x_data))(w_star)
cov = jnp.linalg.inv(-hess)  # Negative Hessian is precision

# Sample from Laplace approximation
S = 100
samples = np.random.multivariate_normal(mean=w_star, cov=np.array(cov), size=S)

# Generate function curves from samples
x_plot = jnp.linspace(0, 10, 300)
f_samples = jnp.array([f(w[0], w[1], x_plot) for w in samples])

# Plot
plt.figure(figsize=(10, 6))

# Plot all sampled functions
for f_s in f_samples:
    plt.plot(x_plot, f_s, color='gray', alpha=0.3, linewidth=1)

# Overlay data
plt.scatter(x_data, y_data, color='red', label='Data', zorder=5)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$f(x)$', fontsize=12)
plt.title(f'{S} Samples from Laplace Approximation', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

**Task 2.8: Use the Laplace approximation to compute the (approximate) posterior probability
for the event $f(8) > 3$.**

```python
S = 10000  # Large number for better estimate
samples = np.random.multivariate_normal(mean=w_star, cov=np.array(cov), size=S)

# Compute f(8) for all samples
x_val = 8.0
f_8_samples = jnp.array([w[1] * sigmoid(x_val + w[0]) for w in samples])

# Estimate P(f(8) > 3)
prob_f8_gt_3 = jnp.mean(f_8_samples > 3)

print(f"Approximate posterior probability that f(8) > 3: {prob_f8_gt_3:.4f}")
```

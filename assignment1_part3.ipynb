$\Huge \textbf{Part 3:}$

In part 3 we will be working with the Poisson-Gamma model:
$$
\begin{align*}
p(y_i|\lambda) &= \frac{\lambda^{y_i}}{y_i!}\exp(-\lambda), \\

p(\lambda|a_0,b_0) &= \frac{b_0^{a_0}}{\Gamma(a_0)}\lambda^{a_0-1}\exp(-b_0\lambda), 
\end{align*}
$$
where $y_i \isin \mathbb{N}_0\ \text{and}\ \lambda,\ a_0,\ b_0 \isin \mathbb{R}^+$.

$\textbf{Task 3.1}$ Determine the joint distribution $p(\mathbf{y},\lambda)$.

Collecting a set of $N$ observations of $y_i$ and assuming each of them are independent conditioned on $\lambda$, the probability of observing $\mathbf{y}=\{ y_i\}^N_{i=1}$ is given by

$$
p(\mathbf{y}|\lambda)=\prod^N_{i=1} p(y_i|\lambda)

$$

The joint distribution can be written as a conditional distribution
$$
\begin{equation*}
p(\mathbf{y},\lambda) = p(\mathbf{y}|\lambda)p(\lambda)
\end{equation*}
$$
Studying the Poisson-Gamma model we get
$$
\begin{align*}
p(\mathbf{y},\lambda|a_0,b_0) &= \prod^N_{i=1} \frac{\lambda^{y_i}}{y_i!}\exp(-\lambda)\frac{b_0^{a_0}}{\Gamma(a_0)}\lambda^{a_0-1}\exp(-b_0\lambda)\\
&=
\frac{b_0^{a_0}}{\Gamma(a_0)} \frac{\lambda^{\left(a_0 - 1 + \sum^N_{i=1}y_i \right)}}{\prod^N_{i=1}y_i!}\exp(-(N+b_0)\lambda)\\
&\stackrel{(*)}{=}
\frac{b_0^{a_0}}{\Gamma(a_0)} \frac{\lambda^{\left(a_0 - 1 + N\bar{y} \right)}}{\prod^N_{i=1}y_i!}\exp(-(N+b_0)\lambda),
\end{align*}
$$
where we have used the average $\bar{y}=\frac 1N\sum_i^N y_i$ in (*) to simplify the expression.

From this we also see that the posterior must have the same form as the prior, since it is independent of $\mathbf{y}$, hence they are conjugate wrt. $p(\mathbf{y}|\lambda)$ as expected.
$$
\begin{equation*}
p(\lambda|\mathbf{y}) \propto  \lambda^{\left(a_0 - 1 + N\bar{y} \right)}\exp(-(N+b_0)\lambda) \propto \text{Gamma}(a_0 + N\bar{y}, b_0 + N)

\end{equation*}
$$


$\textbf{Task 3.2}$ Functional form of Gamma distribution.

For numerical stability, the log transform is often used. For a Gamma distribution the log transform is as follows:
$$
\begin{align*}
\log p(\lambda|a,b) &= \log\left( \frac{b^{a}}{\Gamma(a)}\lambda^{a-1}\exp(-b\lambda)\right)\\
&=
\underbrace{a\log b - \log \Gamma(a)}_{\text{constant}} + (a-1)\log \lambda - b\lambda\\
\end{align*}
$$


$\textbf{Task 3.3}$ Analytical expression for the posterior distribution.

From Bayes theorem we know that
$$
\begin{align*}
p(\lambda|\mathbf{y}) = \frac{p(\mathbf{y}|\lambda)p(\lambda)}{p(\mathbf{y})},
\end{align*}
$$
where $p(y)$ is the evidence, acting as a normalisation factor.
$$
\begin{align*}
p(\mathbf{y}) = \int_{\lambda} p(\mathbf{y}|\lambda)p(\lambda)\ \text{d}\lambda 
\end{align*}
$$

We can then evaluate the intergral of the joint distribution from task 3.1, since it is just a new gamma function
$$
\begin{align*}
p(y) &= \frac{b_0^{a_0}}{\Gamma(a_0)\prod^N_{i=1}y_i!} \int_{\lambda} \lambda^{\left(a_0 - 1 + N\bar{y} \right)}\exp(-(N+b_0)\lambda)     \      \text{d}\lambda \\
&=
\frac{b_0^{a_0}}{\Gamma(a_0)\prod^N_{i=1}y_i!} \frac{\Gamma(a_0 + N\bar{y})}{(b_0+N)^{a_0+N\bar{y}}}.

\end{align*}
$$
To ease the notation, we will denote $a_1 = a_0 + N\bar{y}$ and $b_1 = b_0 + N$.
Normalising the joint distribution found in task 3.1, we can then derive the posterior distribution
$$
\begin{align*}
p(\lambda|\mathbf{y}) &= \frac{\overbrace{p(\mathbf{y}|\lambda)p(\lambda)}^{p(\mathbf{y},\lambda)}} {p(\mathbf{y})}\\
&=
\underbrace{\frac{b_0^{a_0}}{\Gamma(a_0)} \frac{\lambda^{a_1-1}}{\prod^N_{i=1}y_i!}\exp(-b_1\lambda)}_{p(\mathbf{y},\lambda)} \left( \underbrace{\frac{b_0^{a_0}}{\Gamma(a_0)\prod^N_{i=1}y_i!} \frac{\Gamma(a_1)}{(b_1)^{a_1}}}_{p(\mathbf{y})} \right)^{-1}\\
&=
\frac{b_1^{a_1}}{\Gamma(a_1)}\lambda^{a_1-1}\exp(-b_1\lambda)\\
&=
\text{Gamma}(a_0 + N\bar{y}, b_0 + N)
\end{align*}
$$

We see that this is consistent with the conjugate statement about the posterior in task 3.1.
 



$\textbf{Task 3.4}$ Posterior distribution of $\lambda$ given the data: 
$\mathbf{y}_{\text{sample}} = \{y_1=7,y_2=4,y_3=8,y_4=11,y_5=12\}$, as well as assuming $a_0=1$ and $b_0=1/10$. 

The expected value of the posterior is given by
$$
\begin{align*}
\mathbb{E}[\lambda] =& \int_{\lambda} \lambda \frac{b_1^{a_1}}{\Gamma(a_1)}\lambda^{(a_1-1)}\exp(-b_1\lambda) \ \text{d}\lambda\\
=&
\frac{b_1^{a_1}}{\Gamma(a_1)}\int_{\lambda} \lambda^{a_1}\exp(-b_1\lambda) \ \text{d}\lambda\\
=&
\frac{b_1^{a_1}}{\Gamma(a_1)}\frac{\Gamma(a_1+1)}{b_1^{(a_1+1)}}\\
\stackrel{(**)}{=}&
\frac{a_1}{b_1} \\
=& \frac{a_0+N\bar{y}}{b_0+N},
\end{align*}
$$
where in (**) we have used $\Gamma(z+1)=z\Gamma(z)$.

We can then determine the mean of $\lambda$ conditioned on the given data. The posterior distribution, given $\mathbf{y}_{\text{sample}}$, is
$$
p(\lambda|\mathbf{y}_{\text{sample}}) = \text{Gamma}(a_0 + N\bar{y}, b_0 + N)= \text{Gamma}(43,5.1),
$$
and the mean given the data 
$$
\begin{align*}
\mathbb{E}[\lambda] =\frac{a_0+N\bar{y}}{b_0+N}= \frac{43}{5.1} = 8.4
\end{align*}
$$

```python
import jax.numpy as jnp
import matplotlib.pyplot as plt
import seaborn as snb
from scipy.special import gammaln

snb.set_theme(font_scale=1.25)
```

```python
y = jnp.array([7, 4, 8, 11, 12])
N = len(y)
a0, b0 = 1, 0.1

lambda_ = jnp.linspace(0, 30, 1000)

def log_Gamma(lamb, y, a, b):
    
    constant = a*jnp.log(b) - gammaln(a)
    return constant + (a-1)*jnp.log(lamb) - b*lamb

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(lambda_, jnp.exp(log_Gamma(lambda_, y, a0, b0)), label=r'prior, p($\lambda|a_0,b_0$)')
ax.plot(lambda_, jnp.exp(log_Gamma(lambda_, y, a0 + y.sum(), b0 + N)), label=r'posterior, p($\lambda|\mathbf{y}$)')
ax.set_xlabel(r'$\lambda$')
ax.set_ylabel('Probability density')
ax.set_title(r'Prior and posterior distributions: $a_0={}, b_0={}, N={}$'.format(a0, b0, N))
plt.legend()
plt.show()
```

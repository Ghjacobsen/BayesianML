$\textbf{Task 1.3}$ Compute and plot the posterior predictive distribution for $y^*$ given $y = 4$

We know that the posterior predictive distribution is given as

\begin{equation}
p(y^* | y) = \int p(y^* \vert \theta) p(\theta \vert y) d\theta
\end{equation}

We choose to sample different values of $\theta$ from the posterior and then afterwards we sample new predictions $y^*$ from the posterior predictive using the different values of $\theta$

```python

a, b = 5, 112  # Parameters of the posterior Beta distribution
N_star = 20  # Future trials

#We sample from the posterior distribution in order to visualize the posterior
#predictive with the samples we get.
theta_samples = np.random.beta(a, b, 100000)

#Once we have the posterior samples we can sample from the binomial predictive distribution.
y_star_samples = np.random.binomial(N_star, theta_samples)

# Step 3: Compute and plot the posterior predictive distribution
plt.hist(y_star_samples, bins=range(N_star+2), density=True, alpha=0.75, edgecolor='black')
plt.xlabel(r'$y^*$ (Future successes)')
plt.ylabel('Probability')
plt.title(f'Posterior Predictive Distribution for $y^*$ (N* = {N_star})')
plt.xticks(range(N_star+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

We can also compute it analytically. Here $a=y+a_0$ and $b= N-y-b_0$). This can be done computing the integral:
\begin{equation}
p(y^* | y) = \int p(y^* \vert \theta) p(\theta \vert y) d\theta = \binom{N^*}{k}\frac{B(a+k,b+N^*-k)}{B(a,b)}=\binom{20}{k}\frac{B(5+k,132-k)}{B(5,112)}
\end{equation}

```python
def binomial_beta(k, a, b, N):
    return math.comb(N, k) * beta_func(a + k, b + N - k) / beta_func(a, b)

a, b = 5, 112
N_star = 20
k_values = np.arange(0, N_star+1)
post_pred_values = [binomial_beta(k, a, b, N_star) for k in k_values]


plt.plot(k_values, post_pred_values, 'o-', color='black')
plt.xlabel(r'$y^*$ (Future successes)')
plt.ylabel('Probability')
plt.title(f'Posterior Predictive Distribution for $y^*$ (N* = {N_star})')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Part 2

In part 2 we will be working with Linear Gaussian systems.

We let $z_1,z_2 \in \mathbb{R^2}$, and $y \in \mathbb{R}$ be random variables and we consider the following system

\begin{align}
z_1 &\sim \mathcal{N}(0,vI) \tag{1} \\
z_2 \vert z_1 &\sim \mathcal{N}(z_1,vI) \tag{2} \\
y \vert z_2 &\sim \mathcal{N}(a^Tz_2,\sigma^2) \tag{3}
\end{align}

where $a \in \mathbb{R^2}$ is constant. The joint distribution of $(z_1, z_2, y)$ is given by

\begin{equation}
p(y,z_1,z_2) = p(y \vert z_2)p(z_2 \vert z_1) p(z_1) \tag{4}
\end{equation}

Let's try and put some words on the information provided here. For instance, we can see that the distribution of $z_2|z_1$ is conditionally dependent on $z_1$.

We also see that $y$ must be conditionally independent of $z_1$ since if $z_2$ is known, then $z_1$ is not needed, and thus we have independence. This allows us to write:

\begin{equation}
p(y \vert z_1,z_2) = p(y | z_2) \tag{5}
\end{equation}

$\textbf{Task 2.1}$ Determine the distribution $p(y)$.

We start by wanting to compute $p(z_2)$ first, as this allows us to use the following equation:

\begin{equation}
p(y) = \int p(y \vert z_2) p(z_2)dz_2 \tag{6}
\end{equation}

We make use of both the Conditional rule and the Sum rule where we have the following 
\begin{align} 
P(b) &= \int p(a, b)da  \hspace{1cm}
p(a \vert b) &= \frac{p(a, b)}{p(b)} \tag{7}
\end{align}

This means that we can do this manipulation

\begin{align*}
p(z_2 \vert z_1) &= \frac{p(z_2, z_1) }{p(z_1)} \hspace{1cm}
p(z_2, z_1) = p(z_2 \vert z_1) p(z_1) \tag{8}
\end{align*}

This expression can be substituted into the sum rule, and now we want to marginalize out the $z_1$ factor so we are left with the $z_2$ distribution

\begin{equation*}
p(z_2) = \int p(z_2 | z_1) p(z_1) dz_1 \tag{9}
\end{equation*}

Once we are here we can utilize the equations for bayes rules for gaussian given in Murphy (3.37)

\begin{align*}
        p(z_2) &\sim \mathcal{N}(0 + z_1, vI + vI) \\ &\sim \mathcal{N}(z_1, 2vI) \\ &\sim  \mathcal{N}(0,2vI) \tag{10}
\end{align*}

We now complete our equation for $p(y)$ and marginalize out $z_2$ this time:

\begin{equation}
p(y) = \int p(y | z_2)p(z_2) d z_2
\end{equation}

However we see again that $y$ is a linear transformation of a gaussian variable plus gaussian noise. This gives:
$y \sim a^Tz_2+ \epsilon$ where $\epsilon\sim\mathcal{N}(0,\sigma^2)$

It is easy to see that the mean of $y$ becomes 0 as $E[y]=E[a^T z_2] + E[\epsilon]=a^TE[z_2]+0=0$.

With the law of total variance (murphy (2.45)) we get the variance of $y$ as:

\begin{equation}
Var(y)=Var(a^T z_2)+\sigma^2
\end{equation}

Then: 
\begin{equation}
Var(a^T z_2) = a^T Cov(z_2)a=a^T(2vI)a=2v(a^Ta)
\end{equation}

Therefore:
\begin{equation}
Var(y)=2v(a^Ta)+\sigma^2
\end{equation}

This gives us the marginal distribution of y as:

\begin{equation}
p(y) \sim \mathcal{N}(0, 2v(a^Ta)+\sigma^2)
\end{equation}


$\textbf{Task 2.2 }$ Determine the distribution $p(y,z_2 | z_1)$

We use the rules of conditional independence to determine this joint distribution. Lets start by using the chain rule of probability so we can rewrite

\begin{equation*}

p(y,z_2 | z_1) = p(y|z_2,z_1)p(z_2 \vert z_1) \tag{12}

\end{equation*}

We remember our deduction in equation (5) and our expression becomes

\begin{equation*}

p(y,z_2 | z_1) = p(y|z_2)p(z_2 | z_1) \tag{13}

\end{equation*}

When we multiply two gaussians we get a two-dimensional gaussian distribution, so our mean vector will consist of the means of $y$ and $z_2$, and our covariance matrix include the covariance between and the individual variances.

\begin{equation*}

\mathcal{N} \left(\begin{bmatrix} \mathbb{E}(y | z_1) \\ \mathbb{E}(z_2) \end{bmatrix} , \begin{bmatrix} \sigma^2(y|z_1) & \Sigma(y,z_2) \\ \Sigma(y,z_2) & \sigma^2(z_2|z_1) \end{bmatrix} \right) \tag{14}

\end{equation*}

Here we see that the only value we are missing in order to plug in our distribution is the covariance. Covariance between $y$ and $z_2$ is given by the following equation 

\begin{align*}

\Sigma(y, z_2) &= \mathbb{E}[(y-\mathbb{E}[y | z_1])(z_2-\mathbb{E}[z_2  \vert z_1])^T] \\
            &= \mathbb{E}[(a^Tz_2-a^Tz_1)(z_2-z_1)^T] \\
            &= a^T \cdot \mathbb{E}[(z_2 - z_1)(z_2-z_1)^T] \\
            &= a^T \cdot vI \\
            &= va^T \tag{15}
\end{align*}

Lets quickly explain some intuition behind step 2 and three in equation (15). We know that from our model that $y = a^Tz_2$, so when we take the expected value we get 
\begin{equation*}
\mathbb{E}(y | z_1) = \mathbb{E}(a^Tz_2) = a^T\mathbb{E}(z_2) = a^Tz_1 \tag{16}
\end{equation*}

We also take advantage of the fact that $z_2 - z_1$ follow a gaussian noise model with variance $vI$ so we can substitue that in. The uncertainties added in these steps offset each other due to difference in signs. Our final distribution becomes

\begin{equation*}

p(y, z_2 | z_1) \sim \mathcal{N} \left(\begin{bmatrix} a^Tz_1 \\ z_1 \end{bmatrix} , \begin{bmatrix} va^Ta + \sigma^2 & va^T \\ va & vI \end{bmatrix} \right) \tag{17}

\end{equation*}

where the covariance term has to be symmetric so we remove the transpose in the left corner.

$\textbf{Task 2.3}$ Determine the distribution $p(y \vert z_1)$

We have actually already found the distribution, the mean and variance can be extracted from the distribution we just described, by looking at the mean and variance in the vector and matrix.

\begin{equation*}
p(y | z_1) \sim \mathcal{N} (a^Tz_1, \sigma^2 + va^Ta) \tag{18}
\end{equation*}

$\textbf{Task 2.4}$ Determine the distribution $p(z_1 \vert y)$

Lets start by writing up bayes theorem for this distribution given as 

\begin{equation*}
    p(z_1 | y) = \frac{p(y|z_1)p(z_1)}{p(y)} \tag{19}
\end{equation*}

We again use the equations described in Murphy and we see that for the gaussian linear models posterior distribution described as

\begin{equation*}
    p(z_1 | y) \sim \mathcal{N}(\mu_{z_1 | y}, \Sigma_{z_1|y}) \tag{20}
\end{equation*}

We have the following 

\begin{equation}
\Sigma_{z | y}^{-1} = \Sigma_z^{-1} + \mathbf{W}^T \Sigma_y^{-1} \mathbf{W} \tag{21}
\end{equation}

\begin{equation}
\mu_{z | y} = \Sigma_{z | y} \left[ \mathbf{W}^T \Sigma_y^{-1} (y - b) + \Sigma_z^{-1} \mu_z \right] \tag{22}
\end{equation}

We now have to substitue in our know values in these expressions, we start by substituting in our terms for the variance and we end up with

\begin{align*}
\Sigma_{z_1 | y}^{-1} &= (vI)^{-1} + a^T (v a^T a + \sigma^2)^{-1} a, \\
&= \frac{1}{v} I + \frac{a a^T}{v a^T a + \sigma^2}. \tag{23}
\end{align*}

We flip the exponent so we get

\begin{equation*}
\Sigma_{z_1 | y} = \left( \frac{1}{v} I + \frac{a a^T}{v a^T a + \sigma^2} \right)^{-1}. \tag{24}
\end{equation*}

After having found this term we can now substitue our values into the expression for the mean given in equation (22)

\begin{align*}
\mu_{z_1 | y} &= \Sigma_{z_1 | y} \left[ a^T (v a^T a + \sigma^2)^{-1} y + (vI)^{-1} \cdot 0 \right] \\
&= \Sigma_{z_1 | y} \left( \frac{a}{v a^T a + \sigma^2} y \right) \tag{25}
\end{align*}

Leaving us with a final distribution given as

\begin{equation}
p(z_1 | y) \sim \mathcal{N} \left( \frac{v a}{v a^T a + \sigma^2} y, \left( \frac{1}{v} I + \frac{a a^T}{v a^T a + \sigma^2} \right)^{-1} \right). \tag{26}
\end{equation}



## Part 3

In part 3 we will be working with the Poisson-Gamma model:
$$
\begin{align*}
p(y_i|\lambda) &= \frac{\lambda^{y_i}}{y_i!}\exp(-\lambda), \\

p(\lambda|a_0,b_0) &= \frac{b_0^{a_0}}{\Gamma(a_0)}\lambda^{a_0-1}\exp(-b_0\lambda), 
\end{align*}
$$
where $y_i \in \mathbb{N}_0\ \text{and}\ \lambda,\ a_0,\ b_0 \in \mathbb{R}^+$.

$\textbf{Task 3.1}$ Determine the joint distribution $p(\mathbf{y},\lambda)$.

Collecting a set of $N$ observations of $y_i$ and assuming each of them are independent conditioned on $\lambda$, the probability of observing $\mathbf{y}=\{ y_i\}^N_{i=1}$ is given by

$$
p(\mathbf{y}|\lambda)=\prod^N_{i=1} p(y_i|\lambda)

$$

The joint distribution can be written as a conditional distribution
$$
\begin{equation*}
p(\mathbf{y},\lambda) = p(\mathbf{y}|\lambda)p(\lambda)
\end{equation*}
$$
Studying the Poisson-Gamma model we get
$$
\begin{align*}
p(\mathbf{y},\lambda|a_0,b_0) &= \prod^N_{i=1} \frac{\lambda^{y_i}}{y_i!}\exp(-\lambda)\frac{b_0^{a_0}}{\Gamma(a_0)}\lambda^{a_0-1}\exp(-b_0\lambda)\\
&=
\frac{b_0^{a_0}}{\Gamma(a_0)} \frac{\lambda^{\left(a_0 - 1 + \sum^N_{i=1}y_i \right)}}{\prod^N_{i=1}y_i!}\exp(-(N+b_0)\lambda)\\
&\stackrel{(*)}{=}
\frac{b_0^{a_0}}{\Gamma(a_0)} \frac{\lambda^{\left(a_0 - 1 + N\bar{y} \right)}}{\prod^N_{i=1}y_i!}\exp(-(N+b_0)\lambda),
\end{align*}
$$
where we have used the average $\bar{y}=\frac 1N\sum_i^N y_i$ in (*) to simplify the expression.

From this we also see that the posterior must have the same form as the prior, since it is independent of $\mathbf{y}$, hence they are conjugate wrt. $p(\mathbf{y}|\lambda)$ as expected.
$$
\begin{equation*}
p(\lambda|\mathbf{y}) \propto  \lambda^{\left(a_0 - 1 + N\bar{y} \right)}\exp(-(N+b_0)\lambda) \propto \text{Gamma}(a_0 + N\bar{y}, b_0 + N)

\end{equation*}
$$


$\textbf{Task 3.2}$ Functional form of Gamma distribution.

For numerical stability, the log transform is often used. For a Gamma distribution the log transform is as follows:
$$
\begin{align*}
\log p(\lambda|a,b) &= \log\left( \frac{b^{a}}{\Gamma(a)}\lambda^{a-1}\exp(-b\lambda)\right)\\
&=
\underbrace{a\log b - \log \Gamma(a)}_{\text{constant}} + (a-1)\log \lambda - b\lambda\\
\end{align*}
$$


$\textbf{Task 3.3}$ Analytical expression for the posterior distribution.

From Bayes theorem we know that
$$
\begin{align*}
p(\lambda|\mathbf{y}) = \frac{p(\mathbf{y}|\lambda)p(\lambda)}{p(\mathbf{y})},
\end{align*}
$$
where $p(y)$ is the evidence, acting as a normalisation factor.
$$
\begin{align*}
p(\mathbf{y}) = \int_{\lambda} p(\mathbf{y}|\lambda)p(\lambda)\ \text{d}\lambda 
\end{align*}
$$

We can then evaluate the intergral of the joint distribution from task 3.1, since it is just a new gamma function
$$
\begin{align*}
p(y) &= \frac{b_0^{a_0}}{\Gamma(a_0)\prod^N_{i=1}y_i!} \int_{\lambda} \lambda^{\left(a_0 - 1 + N\bar{y} \right)}\exp(-(N+b_0)\lambda)     \      \text{d}\lambda \\
&=
\frac{b_0^{a_0}}{\Gamma(a_0)\prod^N_{i=1}y_i!} \frac{\Gamma(a_0 + N\bar{y})}{(b_0+N)^{a_0+N\bar{y}}}.

\end{align*}
$$
To ease the notation, we will denote $a_1 = a_0 + N\bar{y}$ and $b_1 = b_0 + N$.
Normalising the joint distribution found in task 3.1, we can then derive the posterior distribution
$$
\begin{align*}
p(\lambda|\mathbf{y}) &= \frac{\overbrace{p(\mathbf{y}|\lambda)p(\lambda)}^{p(\mathbf{y},\lambda)}} {p(\mathbf{y})}\\
&=
\underbrace{\frac{b_0^{a_0}}{\Gamma(a_0)} \frac{\lambda^{a_1-1}}{\prod^N_{i=1}y_i!}\exp(-b_1\lambda)}_{p(\mathbf{y},\lambda)} \left( \underbrace{\frac{b_0^{a_0}}{\Gamma(a_0)\prod^N_{i=1}y_i!} \frac{\Gamma(a_1)}{(b_1)^{a_1}}}_{p(\mathbf{y})} \right)^{-1}\\
&=
\frac{b_1^{a_1}}{\Gamma(a_1)}\lambda^{a_1-1}\exp(-b_1\lambda)\\
&=
\text{Gamma}(a_0 + N\bar{y}, b_0 + N)
\end{align*}
$$

We see that this is consistent with the conjugate statement about the posterior in task 3.1.
 



$\textbf{Task 3.4}$ Posterior distribution of $\lambda$ given the data: 
$\mathbf{y}_{\text{sample}} = \{y_1=7,y_2=4,y_3=8,y_4=11,y_5=12\}$, as well as assuming $a_0=1$ and $b_0=1/10$. 

The expected value of the posterior is given by
$$
\begin{align*}
\mathbb{E}[\lambda] =& \int_{\lambda} \lambda \frac{b_1^{a_1}}{\Gamma(a_1)}\lambda^{(a_1-1)}\exp(-b_1\lambda) \ \text{d}\lambda\\
=&
\frac{b_1^{a_1}}{\Gamma(a_1)}\int_{\lambda} \lambda^{a_1}\exp(-b_1\lambda) \ \text{d}\lambda\\
=&
\frac{b_1^{a_1}}{\Gamma(a_1)}\frac{\Gamma(a_1+1)}{b_1^{(a_1+1)}}\\
\stackrel{(**)}{=}&
\frac{a_1}{b_1} \\
=& \frac{a_0+N\bar{y}}{b_0+N},
\end{align*}
$$
where in (**) we have used $\Gamma(z+1)=z\Gamma(z)$.

We can then determine the mean of $\lambda$ conditioned on the given data. The posterior distribution, given $\mathbf{y}_{\text{sample}}$, is
$$
p(\lambda|\mathbf{y}_{\text{sample}}) = \text{Gamma}(a_0 + N\bar{y}, b_0 + N)= \text{Gamma}(43,5.1),
$$
and the mean given the data 
$$
\begin{align*}
\mathbb{E}[\lambda] =\frac{a_0+N\bar{y}}{b_0+N}= \frac{43}{5.1} = 8.4
\end{align*}
$$




$\textbf{Task 3.5}$

```python
y = jnp.array([7, 4, 8, 11, 12])
N = len(y)
a0, b0 = 1, 0.1

lambda_ = jnp.linspace(0, 30, 1000)


def log_Gamma(lamb, y, a, b):

    constant = a * jnp.log(b) - gammaln(a)
    return constant + (a - 1) * jnp.log(lamb) - b * lamb


fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(
    lambda_,
    jnp.exp(log_Gamma(lambda_, y, a0, b0)),
    label=r"prior, p($\lambda|a_0,b_0$)",
)
ax.plot(
    lambda_,
    jnp.exp(log_Gamma(lambda_, y, a0 + y.sum(), b0 + N)),
    label=r"posterior, p($\lambda|\mathbf{y}$)",
)
ax.set_xlabel(r"$\lambda$")
ax.set_ylabel("Probability density")
ax.set_title(
    r"Prior and posterior distributions: $a_0={}, b_0={}, N={}$".format(a0, b0, N)
)
plt.legend()
plt.show()
```
